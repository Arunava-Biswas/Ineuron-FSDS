#### Question 1:

##### Write a program that calculates and prints the value according to the given formula:

- `Q = Square root of [(2 * C * D)/H]`

- Following are the fixed values of C and H:

- `C is 50. H is 30`

- `D is the variable whose values should be input to your program in a comma-separated sequence`

- **Example**

- Let us assume the following comma separated input sequence is given to the program:

- 100,150,180

- The output of the program should be:

- 18,22,24


```python
# importing module
from math import sqrt

# user input
num = eval(input("Enter the input in comma separated sequence: "))

# defining function
def calculate_squareRoot(n):
    output = []
    C = 50 # Declaring and initializing constant C
    H = 30 # Declaring and initializing constant H
    for D in n:
        Q = str(int(sqrt((2 * C * D)/ H)))
        output.append(Q)
    return (','.join(output))

# calling the function:
output = calculate_squareRoot(num)
print(f"Output: {output}")
```

#### Question 2:

##### Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The element value in the i-th row and j-th column of the array should be i*j.

- `Note: i=0,1.., X-1; j=0,1,¡¬Y-1`

- **Example**

- Suppose the following inputs are given to the program:

- `3,5`

- Then, the output of the program should be:

- `[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]] `


```python
# importing module
import array as arr

# taking user inputs
X = int(input("Enter number of rows: "))
Y = int(input("Enter number of columns: "))

# defining the function
def array_generation(X, Y):
    output = []
    for i in range(X):
        output.insert(X,[])
        for j in range(Y):
            output[i].append(i * j)
    return output

# calling the function
Arr = array_generation(X, Y)
print(Arr)
```

#### Question 3:

##### Write a program that accepts a comma separated sequence of words as input and prints the words in a comma-separated sequence after sorting them alphabetically.

Suppose the following input is supplied to the program:

- `without,hello,bag,world`

Then, the output should be:

- `bag,hello,without,world`

```python
# taking user inputs
input_str = input("Enter the string: ")


# defining function
def stringSorting(s):
    split_string = s.split(',')   # splitting the string using separator ','
    output = ','.join(sorted(split_string))   # joining the string using separator ',' after sorting it
    return output

# calling the function
final_str = stringSorting(input_str)
print("Output: ", final_str)
```

#### Question 4:

##### Write a program that accepts a sequence of whitespace separated words as input and prints the words after removing all duplicate words and sorting them alphanumerically.

Suppose the following input is supplied to the program:

- `hello world and practice makes perfect and hello world again`

Then, the output should be:

- `again and hello makes perfect practice world`

```python
# taking user inputs
input_str = input("Enter the string: ")


# defining function
def alnum_Sorting(s):
    split_string = s.split(' ')   # splitting the string using separator ' '
    # joining the string using separator ' ' after sorting it and removing duplicates
    output = ' '.join(sorted(sorted(list(set(split_string)))))   
    return output

# calling the function
final_str = alnum_Sorting(input_str)
print("Output: ", final_str)
```

#### Question 5:

##### Write a program that accepts a sentence and calculate the number of letters and digits.

Suppose the following input is supplied to the program:

- `hello world! 123`

Then, the output should be:

- `LETTERS 10`

- `DIGITS 3`


```python
# taking user inputs
input_str = input("Enter the string: ")

# defining function
def letter_digit_counting(string):
    lettersList = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz'
    digitsList = '0123456789'
    letters_count = 0
    digits_count = 0
    
    for i in string:
        if i in lettersList:
            letters_count += 1
        if i in digitsList:
            digits_count += 1
    return letters_count, digits_count


# calling the function
letters, digits = letter_digit_counting(input_str)
print(f"LETTERS {letters} \nDIGITS {digits}")
```

#### Question 6:

##### A website requires the users to input username and password to register. Write a program to check the validity of password input by users.

Following are the criteria for checking the password:

1. `At least 1 letter between [a-z]`

2. `At least 1 number between [0-9]`

3. `At least 1 letter between [A-Z]`

4. `At least 1 character from [$#@]`

5. `Minimum length of transaction password: 6`

6. `Maximum length of transaction password: 12`

Your program should accept a sequence of comma separated passwords and will check them according to the above criteria. Passwords that match the criteria are to be printed, each separated by a comma.

- **Example**

If the following passwords are given as input to the program:

- `ABd1234@1,a F1#,2w3E*,2We3345`

Then, the output of the program should be:

- `ABd1234@1`


```python
# taking user inputs
input_pass = input("Enter the passwords separated by comma: ")


# defining function
def passwordCheck(string):

    num_list = "0123456789"
    special_list = "$#@"
    
    for k in string.split(","):
        if len(k) <= 12 and len(k) >= 6:
            if any(i.isupper() for i in k):
                if any(i.islower() for i in k):
                    if any(i for i in k if i in num_list):
                        if any(i for i in k if i in special_list):
                            out = k                    
    return out


# calling the function
output = passwordCheck(input_pass)
print(output)
```
