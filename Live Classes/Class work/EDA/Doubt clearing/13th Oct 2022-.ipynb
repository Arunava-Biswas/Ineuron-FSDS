### 2. `mean`,` median`,`mode` method

#### With `mean`

```python
# Again create a copy of the data

df_copy = df.copy()
df_copy.shape
```

```python
# Here we perform with the Size column

df_copy['Size'].isnull().sum()
```

```python
# Now we will fill the missing values with the mean
# For this we will use the fillna()

df_copy['Size'] = df_copy['Size'].fillna(df_copy['Size'].mean())
df_copy['Size']
```

```python
# Now again checking for null values in the same column

df_copy['Size'].isnull().sum()
```

- We can perform the same with `median` and `mode` also.

```python
# Now check the distribution of the data

plt.figure(figsize=(18,8))

plt.subplot(1, 2, 1)
plt.title('Distribution With Null Values')
sns.distplot(df["Size"])

plt.subplot(1, 2, 2)
plt.title('Distribution With mean Values')
sns.distplot(df_copy["Size"])

plt.show()
```

#### With `median `

```python
# Creating another copy

df_copy = df.copy()
df_copy.shape
```

```python
# Now changing the missing values with median

df_copy['Size'] = df_copy['Size'].fillna(df_copy['Size'].median())
df_copy['Size']
```

```python
# Now check the distribution of the data

plt.figure(figsize=(18,8))

plt.subplot(1, 2, 1)
plt.title('Distribution With Null Values')
sns.distplot(df["Size"])

plt.subplot(1, 2, 2)
plt.title('Distribution With median Values')
sns.distplot(df_copy["Size"])

plt.show()
```

**Observations**

- So we can see that median has a better distribution.

### 3. With `Random sample imputation`


```python
# Creating another copy

df_random = df.copy()
df_random.shape
```

```python
# Here again we will perform on the Size column

df_random['Size'].isnull().sum() 
```

**Notes**

- Now we will take random samples of that many numbers as much we have to replace as here the number is 1526.
- But remember before taking the samples we need to drop the null values or else it may take a sample with null value.

```python
# Now we will take random samples after dropping the null values

df_random["Size"].dropna().sample(1526)
```

```python
# Now we can pass the number of null values inplace of the hardcore number
# As here we are using df_random['Size'].isnull().sum() inplace of 1526
# And then we can take the entire thing into a variable

random_sample = df_random["Size"].dropna().sample(df_random['Size'].isnull().sum())
random_sample
```

```python
# Now fill the missing values with the random values
# Here we are using the index to get the indexes with the null values
# we are storing them in the index of random_sample variable

random_sample.index = df_random[df_random['Size'].isnull()].index
random_sample.index
```

```python
# Now with the help of loc() we are going to replace the null values

df_random.loc[df_random["Size"].isnull(), "Size"] = random_sample
```

```python
# Checking the number of null values again

df_random['Size'].isnull().sum() 
```

```python
# Now check the distribution of the data

plt.figure(figsize=(18,8))

plt.subplot(1, 2, 1)
plt.title('Distribution With Null Values')
sns.distplot(df["Size"])

plt.subplot(1, 2, 2)
plt.title('Distribution With Random Values')
sns.distplot(df_random["Size"])

plt.show()
```
