### 13. Total contributation hour for each and every agents weekly basis

```python
df1 = pd.read_csv('dataset/agent_login.csv', index_col='SL No')
df1.head()
```

```python
# creating a new dataframe with the required columns

df = df1[['Agent', 'Date', 'Duration']]

df.head()
```

```python
# change the Date to datetime and keep it in a new column

date = pd.to_datetime(df['Date'], dayfirst=True)

df['Date'] = date

df.head()
```

```python
# changing the duration to timedelta

df['Duration'] = pd.to_timedelta(df['Duration'])

df.head()
```

```python
df.dtypes
```

```python
# 1st group the data by datewise

df3 = df.groupby(['Agent','Date'])['Duration'].sum().reset_index()

df3.rename(columns={'Duration' : 'Total_time'}, inplace=True)

df3.head()
```

```python
# Now create week column

df3['week'] = df3['Date'].dt.week

df3.head()
```

```python
# Now groupby weekwise to get the answer

df4 = df3.groupby(['Agent', 'week'])['Total_time'].sum().reset_index()

df4.head(20)
```

### 14. total percentage of active hour for a month

```python
df4.head()
```

```python
print(f"Total active hour in the month is: {df4['Total_time'].sum()}")
```

```python
df4['Total Active Hour'] = df4['Total_time'].sum()

df4.head()
```

```python
def perc_hr(t):
    active_hour = df4['Total_time'].sum()
    active_per = (t / active_hour) * 100
    return active_per

df4['Active_percentage'] = df4['Total_time'].apply(perc_hr)

df4.head(20)
```
