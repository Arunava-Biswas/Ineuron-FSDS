#### Feature Scaling / Standardization:


- In Linear Regression we get Gradient Descent. So with the help of **Feature scaling (Standardization)** we can reach to the Global minima at a faster rate.
- So here we will use the Standardization i.e. **[(X - ) / ]**.
- This is the 1st step of Feature Engineering.
- Remember in `Decision Tree`, `XG-Boost`, `Random Forest` we do not need to do the Feature Scaling.
- In StandardScaler the ** = 0** and ** = 1**.

```python
## Standardize or feature scaling the datasets

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Now doing fit_transform() the training dataset. Here it will first fit the data and then transform it.
# That is it will compute the  of the data points and then use  to create new data points on same scale. 
# It is the using Z score formula.

X_train = scaler.fit_transform(X_train)
X_train
```

```python
# Here we do the transform() the test data. It will perform standardization by centering and scaling.

X_test = scaler.transform(X_test)
X_test
```

**Notes**

-  To avoid data lickage we use `transform()`. That is the model should not know anything other than the training data.
- So whatever techniques are applied to the training data the same will be applied to the test data.
- But if we do `fit_transform()` then the model will again calculate the **** and **** for the test dataset, and as a result it will get to know about the data points of the test dataset.
- This concept is called **Data Lickage**, that is not to lick any test data to the model during training time.
- Example: The test dataset is like the examination and the train dataset is the syllabus. So we should not know the questions of the exam during our preparation of the exam.

### Model Training:

- First import the model.
- Then create an object.
- Now train the model with the training dataset. We can also achieve the same Standardization by passing the parameter `normalize=True` to the LinearRegression(). But now it is deprecated.
- `fit()` basically means training the data. It is only to find the `coefficient`.
- When we used `fit_transform()` in **StandardScaler**, it was beacause we want to apply one formula to all the variables and then change the data. So `fit_transform()` is basically to send the data and then change the data.

```python
# As it is a multiple Linear Regression model as there are many input features

from sklearn.linear_model import LinearRegression
```

```python
# Creating the object

regression = LinearRegression()
regression
```

```python
# training with data both independent and dependent feature, for this we use fit()

regression.fit(X_train, y_train)
```

```python
# print the coefficients and the intercept
# to get coefficient we use '.coef_' and for intercept we use '.intercept_'

print(regression.coef_)
```

```python
print(regression.intercept_)
```

- So it means that if all the independent features of the model are 0 i.e. 'mx' becomes 0 then the price of the house will be 22 i.e. the 'c'.

```python
# Doing prediction for the test data


reg_pred = regression.predict(X_test)
reg_pred
```

### Assumptions Of Linear Regression


- To check whether the Linear model is good or bad.
- 1st to do create a scatter plot between the real test data and predicted test data of the model.
- 2nd get the residuals or errors and then create a distribution plot of those residuals.
- 3rd create an Uniform distribution by using a scatter plot between the predictions and the residuals.

```python
plt.scatter(y_test, reg_pred)
plt.xlabel("Original Test Data points")
plt.ylabel("Predicted Test Data points")
```

- Here when use scatter plot we get a linear relation between the actual test data and the predicted test data, so the model is good.
- There are still outlier there in the dataset.

```python
# Calculating the residuals, i.e. the errors
# residuals will be the (actual data point - predicted data point)

residuals = y_test - reg_pred
residuals
```

```python
# let's plot the residuals with a Kernal Density Estimator

sns.displot(residuals, kind="kde")
```

- It gives a Gausiian Distribution Graph with a little Right Skewedness. It happens because of the outliers. We will remove these outliers using the Ridge and Lasso.
- So whenever we want to create a Distribution plot of residuals for a Linear regression model we should be able to get almost a normal distribution.

```python
# Scatter plot with predictions and residual (Uniform Distribution)

plt.scatter(reg_pred, residuals)
```

- Uniform Distribution is an important one.
- Here the data points are just randomly distributed, i.e. it does not have any shape.
- It also provides an assumption that the model is good.
- This is the **Homoscedasticity**. i.e. an assumption of equal or similar variances in different groups being compared.

### Let's find the Performance metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("The Mean Squared Error is: ", mean_squared_error(y_test, reg_pred))
print("The Mean Absolute Error is: ", mean_absolute_error(y_test, reg_pred))
print("The Root Mean Squared Error is: ", np.sqrt(mean_squared_error(y_test, reg_pred)))
```

- Here we are finding the loss values.

### R Squared and Adjusted R Square

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, reg_pred)
print("The R Square value is: ", score)
```

```python
adj = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print("The Adjusted R Square value is: ", adj)
```

- So here we can see that Adjusted R square is just a bit less than R squared.

```python
# Creting function for all regreesion models for their errors

def errors(mla, pred_data):
    print(f"For the {mla} algorithm:")
    print("The Mean Squared Error is: ", mean_squared_error(y_test, pred_data))
    print("The Mean Absolute Error is: ", mean_absolute_error(y_test, pred_data))
    print("The Root Mean Squared Error is: ", np.sqrt(mean_squared_error(y_test, pred_data)))
    print("\n")
    
  
errors("Simple Linear Regression", reg_pred)
errors("Ridge Regression", ridge_pred)
errors("Lasso Regression", lasso_pred)
errors("ElasticNet Regression", elastic_pred)
```

```python
# Creting function for all regreesion models for their r values

def r_value(mla, pred_data):
    score = r2_score(y_test, pred_data)
    print(f"The R Square value of {mla} is: {score}")
    adj = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
    print(f"The Adjusted R Square value of {mla} is: {adj}")
    print("\n")
    

r_value("Simple Linear Regression", reg_pred)
r_value("Ridge Regression", ridge_pred)
r_value("Lasso Regression", lasso_pred)
r_value("ElasticNet Regression", elastic_pred)
```
